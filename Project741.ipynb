### Further data preparation

The 'helpful' attribute consisted of entries that had values like [0,0]. Meaning, there were no upvotes or downvotes. These entries weren't useful for our study. Thus, we decided to removie such entires.
The code below keeps entries in a new dataframe where the sum of upvotes and downvotes are greater than 5 i.e <br>
upvotes + downvotes > 5 <br>


```python
df2 = df[df['new_0'] + df['new_1'] > 5]
df2['helpful_percent']=df2['new_0']*100/(df2['new_0']+df2['new_1'])
df2['helpful_percent']=df2['helpful_percent'].astype(int)
```

### Sentiment Analysis Using LIWC

```python
#importing relevant libraries
import liwc
from collections import Counter
```

```python
# read in liwc dictionary and define liwc parser
parse, category_names = liwc.load_token_parser('LIWC2007_English080730.dic')

df2 = pd.DataFrame()

df_final=df_final[~df_final['reviewText'].isnull()]
#creating unique primary key based on index
df_final.reset_index(inplace=True)
df_final
```

```python
# tokenize and pass to liwc
#cp = 0
for index,row in df_final.iterrows(): # loop over rows of data frame
    text = row['reviewText']
    #print(row['asin']) # print name you gave to each text
    #print(row['reviewText'])
    textwords = nltk.tokenize.word_tokenize(text) # tokenize
    
    # get word count
    wc = len(textwords)
    #print("Word count: ",wc)
    
    # now flatmap over all the categories in all of the tokens using a generator:
    text_counts = Counter(category for token in textwords for category in parse(token))
    # for each relevant category print the percentage    
    df1 = pd.DataFrame.from_dict(text_counts, orient='index')
    df1 = df1.T
    df1['rid'] = row['reviewerID']
    df1['pid'] = row['asin']
    df1['helpfulness'] = row['helpful_percent']
    df2 = df2.append(df1)

df2=df2.reset_index(drop=True)
#print(cp)
    
```
